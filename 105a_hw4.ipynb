```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.animation import FuncAnimation
```

```python
j0 = 1
cs = 1
omega0 = 1
qx = 2
```

```python
x = np.linspace(-5, 5, 200)
y = np.linspace(-5, 5, 200)
X, Y = np.meshgrid(x, y)
```

```python
def phi(t):
    return (j0 / (2 * cs * omega0)) * np.cos(omega0 * t) * np.cos(qx * X) * np.exp(-qx * np.abs(Y))
```

```python
fig, ax = plt.subplots()
im = ax.imshow(phi(0), extent=[x.min(), x.max(), y.min(), y.max()])
```

```python
def animate(frame):
    t = frame
    im.set_array(phi(t))
    return [im]
```

```python
t_vals = np.linspace(0, 10, 100)
```

```python
animation = FuncAnimation(fig, animate, frames = t_vals, interval = 60)
plt.show()
```

```python
animation.save("phi.gif", writer = "pillow", fps = 20)
```
