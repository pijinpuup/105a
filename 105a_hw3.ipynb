```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
j = 1j
```

```python
x = np.linspace(-np.pi, np.pi, 1000)
```

```python
def f(x):
    y = np.zeros_like(x)
    maskp = (x > 0) & (x < np.pi)
    maskn = (x < 0) & (x > -np.pi)
    y[maskp] = x[maskp]
    y[maskn] = 0
    return y
```

```python
def c(n: int):
    if n == 0:
        return np.pi / 4
    return 1/(2*np.pi) * (((j * np.pi * (-1) ** n) / n) + (((-1)**n - 1) / n**2))
```

```python
def fN(x, N: int):
    n = np.arange(-N, N+1)
    cn = np.array([c(int(k)) for k in n])
    return np.sum(cn[:, None] * np.exp(j * n[:, None] * x[None, :]), axis = 0)
```

```python
Ns = [0, 1, 2, 3, 4, 5]

plt.figure(figsize = (10, 6))

plt.plot(x, f(x), label = "f(x)", color = 'black', linewidth = 2)

for N in Ns:
    yN = np.real(fN(x, N))
    plt.plot(x, yN, label = f"N={N}", alpha = 0.6)

plt.legend()
plt.show()
```

```python
def df(x):
    y = np.zeros_like(x)
    maskp = (x > 0) & (x < np.pi)
    maskn = (x < 0) & (x > -np.pi)
    y[maskp] = 1
    y[maskn] = 0
    return y
```

```python
def dc(n: int):
    if n == 0:
        return 1/2
    return (((-1)**n) / 2) - ((j / (2*np.pi)) * ((((-1)**n) - 1) / n))
```

```python
def dfN(x, N: int):
    n = np.arange(-N, N+1)
    cn = np.array([dc(int(k)) for k in n])
    return np.sum(cn[:, None] * np.exp(j * n[:, None] * x[None, :]), axis = 0)
```

```python
dNs = [1, 2, 3, 4, 5]

plt.figure(figsize = (10, 6))

plt.plot(x, df(x), label = "df(x)", color = 'black', linewidth = 2)

for dN in dNs:
    dyN = np.real(dfN(x, dN))
    plt.plot(x, dyN, label = f"N={dN}", alpha = 0.6)

plt.legend()
plt.show()
```
